# 02 — Local Tournament

Run local tournaments to measure agent win rate before submitting.

```python
import importlib.util
from pathlib import Path

def load_agent(path='../submission/main.py'):
    spec = importlib.util.spec_from_file_location('agent', path)
    mod = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(mod)
    fns = [(k,v) for k,v in vars(mod).items() if callable(v) and not k.startswith('_')]
    return max(fns, key=lambda kv: getattr(getattr(kv[1],'__code__',None),'co_firstlineno',0))[1]

agent_fn = load_agent()
print('Agent loaded:', agent_fn.__name__)
```

```python
# Quick sanity test
obs = {
    'step': 0, 'player': 1,
    'planets': [
        {'x': 10.0, 'y': 10.0, 'owner': 1, 'ships': 20.0, 'growth_rate': 2.0},
        {'x': 90.0, 'y': 90.0, 'owner': 2, 'ships': 10.0, 'growth_rate': 1.0},
    ],
    'fleets': []
}
action = agent_fn(obs)
print('Action:', action)
```
